```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic 2D data for demonstration
np.random.seed(42)
X = np.random.rand(100, 2) * [100, 200]  # Data with different ranges for each feature

# K-means without normalization
kmeans_no_norm = KMeans(n_clusters=3, random_state=42).fit(X)
labels_no_norm = kmeans_no_norm.labels_

# Normalize data to [0, 1] range
scaler = MinMaxScaler()
X_normalized = scaler.fit_transform(X)

# K-means with normalization
kmeans_norm = KMeans(n_clusters=3, random_state=42).fit(X_normalized)
labels_norm = kmeans_norm.labels_

# Create subplots for visual comparison
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# Plotting K-means without normalization
ax[0].scatter(X[:, 0], X[:, 1], c=labels_no_norm, cmap='viridis', alpha=0.6)
ax[0].scatter(kmeans_no_norm.cluster_centers_[:, 0], kmeans_no_norm.cluster_centers_[:, 1],
              color='red', marker='x', s=100, label='Cluster Centers')
ax[0].set_title('K-means Without Normalization')
ax[0].set_xlabel('Feature 1')
ax[0].set_ylabel('Feature 2')
ax[0].legend()
ax[0].grid(True)

# Plotting K-means with normalization
ax[1].scatter(X_normalized[:, 0], X_normalized[:, 1], c=labels_norm, cmap='viridis', alpha=0.6)
ax[1].scatter(kmeans_norm.cluster_centers_[:, 0], kmeans_norm.cluster_centers_[:, 1],
              color='red', marker='x', s=100, label='Cluster Centers')
ax[1].set_title('K-means With Normalization')
ax[1].set_xlabel('Feature 1 (Normalized)')
ax[1].set_ylabel('Feature 2 (Normalized)')
ax[1].legend()
ax[1].grid(True)

plt.tight_layout()
plt.show()
```
